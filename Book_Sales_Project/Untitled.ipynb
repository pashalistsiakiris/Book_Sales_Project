```python
#now we will create a weighted rating score based on the BAYESIAN average
c=df['Book_average_rating'].mean()
m=1#the minimun number of rating count to be condsired
df['WR']=((df['Book_ratings_count'])/(df['Book_ratings_count']+m)*df['Book_average_rating'])+(m/(m+df['Book_ratings_count'])*c)
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(5,3))
df_author_comp=df.groupby('Author_Rating')['WR'].mean().sort_values(ascending=False).reset_index()
ax=sns.barplot(data=df_author_comp,x='Author_Rating',y='WR')
for index,row in df_author_comp.iterrows():
    ax.text(index,row.WR,round(row.WR,2),ha='center')
plt.title('avg rating per author rating')
    
```
